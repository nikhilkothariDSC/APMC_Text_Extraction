```python
from PIL import Image
 
file = '/kaggle/input/karanja-apmc-market-rates/WhatsApp Image 2023-08-28 at 13.17.07.jpeg'
 
# creating a object
im = Image.open(file)
```

```python
imagetext = pytesseract.image_to_string(im, lang='mar')
print(imagetext)
```

```python
imagetext = pytesseract.image_to_string(im, lang='mar')
soya= (imagetext[imagetext.index('सोया'):imagetext.index('सोया')+27])
from translate import Translator
translator= Translator(from_lang="mr-IN",to_lang="en")
date= (imagetext[imagetext.index('दिनांक')+7:imagetext.index('दिनांक')+17])
soyabean = translator.translate(soya)
date_string= translator.translate(date)
date = datetime.strptime(date_string, '%d/%m/%Y').date()
quantity_market= int(soyabean[-4:])
Avg_price_soya= int(soyabean[14:18])
print(quantity_market)
print(Avg_price_soya)
print(date)
```
